## **Executive Summary**

This report provides an in-depth analysis of U.S. healthcare debt at the county, state, and national levels from 2011 through 2023, examining key demographic and financial factors influencing healthcare debt. Utilizing multiple data sources, including the Urban Institute Credit Profile Board, the American Hospital Association's Annual Survey database, the CMS Provider File, and the American Community Survey, this study evaluates the impact of hospital market concentration, hospital closures and mergers, the uninsured population, the percentage of adults with disabilities, and average household income on medical debt. ### **Key Findings**

- **Association of insurance coverage with medical debt**:  
    - The average uninsured rate in the high medical debt group was 13.42%, significantly higher than the 9.98% rate in the low medical debt group, suggesting that lack of health insurance is a major contributor to increased medical debt.
   
- **Negative correlation between income level and medical debt**:  
    - Low-income counties (<50k) had the highest percentage of medical debt at 27.28%, middle-income counties (50k-75k) had 22.63%, and high-income counties (>75k) had the lowest at 15.60%.
    - This shows that the higher the household income level, the lower the medical debt burden.
   
- **Impact of Insurance Status on Different Income Groups**:  
    - The median medical debt of the uninsured is higher than that of the insured in most income tiers, especially in the middle-income (50k-75k) and high-income (>75k) groups, where the median medical debt reaches 1,013 for middle-income uninsured and 1,007 for high-income uninsured.
    - However, the median medical debt of low-income (<50k) uninsured is slightly lower than that of the insured.
   
- **Racial Disparities**:  
    - People of color groups have a higher percentage of medical debt than white groups, suggesting that people of color are more likely to fall into medical debt, while white groups are less affected.
    - This reflects racial inequalities in the healthcare system.

- **Impact of Hospital Market Concentration**:  
    - Counties with higher hospital market concentration (HHI), such as Gregg County, Texas, show higher levels of medical debt.
    - This suggests that highly concentrated healthcare markets may lead to higher prices for healthcare services, thereby increasing the burden of medical debt on patients.

## __Attribute Information__:

__Year__: The year in which the data was collected or reported for each county.

__County Fips__: A unique identifier code for each county, defined by the Federal Information Processing Standards (FIPS).

__State Abbreviation__: The standard two-letter postal abbreviation of the U.S. state where the county is located.

__County Name__: The name of the county for which the data is recorded.

__Share with medical debt in collections__: The percentage of residents within the county who have medical debt currently in collections.

__Median medical debt in collections in 2023__: The median amount of medical debt in collections for individuals in the county as of 2023.

__Share with medical debt in collections - Majority White__: The percentage of residents identifying as "Majority White" in the county who have medical debt in collections.

__Median medical debt in collections in 2023 - Majority White__: The median medical debt in collections for individuals identifying as "Majority White" in the county, as of 2023.

__Share with medical debt in collections - Majority of Color__: The percentage of residents identifying as "Majority of Color" in the county who have medical debt in collections.

__Median medical debt in collections in 2023 - Majority of Color__: The median medical debt in collections for individuals identifying as "Majority of Color" in the county, as of 2023.

__Hospital market concentration HHI__: The Herfindahl-Hirschman Index (HHI) for hospital market concentration in the county, indicating the level of competition among hospitals (higher values suggest lower competition).

__Number of Closures and Mergers__: The total number of hospital closures and mergers within the county.

__Share of the population with no health insurance coverage__: The percentage of the county's population without any form of health insurance coverage.

__Share of non-elderly adults with a reported disability__: The percentage of non-elderly adults in the county who have reported a disability.

__Average household income in 2023__: The average household income within the county as of 2023.

Calculate the median values for each column by State Abbreviation and store in a temporary table

```python
%%bigquery
CREATE OR REPLACE TABLE `ba775-a07-team.Medical_Debt_Default_Risk_Analysis.median_values` AS
SELECT
  `State Abbreviation`,
  APPROX_QUANTILES(CAST(NULLIF(`Share with medical debt in collections`, 'NA') AS FLOAT64), 100)[OFFSET(50)] AS median_share_with_medical_debt,
  APPROX_QUANTILES(CAST(NULLIF(`Median medical debt in collections in _2023`, 'NA') AS FLOAT64), 100)[OFFSET(50)] AS median_median_medical_debt_in_2023,
  APPROX_QUANTILES(CAST(NULLIF(`Share with medical debt in collections - Majority White`, 'NA') AS FLOAT64), 100)[OFFSET(50)] AS median_share_with_med_debt_majority_white,
  APPROX_QUANTILES(CAST(NULLIF(`Median medical debt in collections in _2023 - Majority White`, 'NA') AS FLOAT64), 100)[OFFSET(50)] AS median_median_medical_debt_in_2023_majority_white,
  APPROX_QUANTILES(CAST(NULLIF(`Share with medical debt in collections - Majority of Color`, 'NA') AS FLOAT64), 100)[OFFSET(50)] AS median_share_with_med_debt_majority_of_color,
  APPROX_QUANTILES(CAST(NULLIF(`Median medical debt in collections in _2023 - Majority of Color`, 'NA') AS FLOAT64), 100)[OFFSET(50)] AS median_median_medical_debt_in_2023_majority_of_color,
  APPROX_QUANTILES(CAST(NULLIF(`Hospital market concentration _HHI_`, 'NA') AS FLOAT64), 100)[OFFSET(50)] AS median_hospital_market_concentration,
  APPROX_QUANTILES(CAST(NULLIF(`Number of Closures and Mergers`, 'NA') AS FLOAT64), 100)[OFFSET(50)] AS median_number_of_closures_and_mergers,
  APPROX_QUANTILES(CAST(NULLIF(`Share of the population with no health insurance coverage`, 'NA') AS FLOAT64), 100)[OFFSET(50)] AS median_share_of_population_no_health_insurance,
  APPROX_QUANTILES(CAST(NULLIF(`Share of non-elderly adults with a reported disability`, 'NA') AS FLOAT64), 100)[OFFSET(50)] AS median_share_nonelderly_adults_disability,
  APPROX_QUANTILES(CAST(NULLIF(`Average household income in _2023`, 'NA') AS FLOAT64), 100)[OFFSET(50)] AS median_average_household_income_2023
FROM `ba775-a07-team.Medical_Debt_Default_Risk_Analysis.changing_med_debt_landscape_county_partial_cleaned`
GROUP BY `State Abbreviation`;
```

__1.5 Updating Missing Values with State-Level Medians__

This query updates the data table by filling missing values in several columns with state-level medians, stored in the median_values table. For each specified column, the query checks if a value is "NA." If it is, it replaces it with the median for that state, retrieved from median_values. This process ensures that missing values are filled with relevant state-specific medians, improving data completeness while maintaining a regional context.

```python
%%bigquery
UPDATE `ba775-a07-team.Medical_Debt_Default_Risk_Analysis.changing_med_debt_landscape_county_partial_cleaned` AS D1
SET
  updated_share_with_medical_debt = IF(D1.`Share with medical debt in collections` = 'NA', median_values.median_share_with_medical_debt, updated_share_with_medical_debt),
  updated_median_medical_debt_in_2023 = IF(D1.`Median medical debt in collections in _2023` = 'NA', median_values.median_median_medical_debt_in_2023, updated_median_medical_debt_in_2023),
  updated_share_with_med_debt_majority_white = IF(D1.`Share with medical debt in collections - Majority White` = 'NA', median_values.median_share_with_med_debt_majority_white, updated_share_with_med_debt_majority_white),
  updated_median_medical_debt_in_2023_majority_white = IF(D1.`Median medical debt in collections in _2023 - Majority White` = 'NA', median_values.median_median_medical_debt_in_2023_majority_white, updated_median_medical_debt_in_2023_majority_white),
  updated_share_with_med_debt_majority_of_color = IF(D1.`Share with medical debt in collections - Majority of Color` = 'NA', median_values.median_share_with_med_debt_majority_of_color, updated_share_with_med_debt_majority_of_color),
  updated_median_medical_debt_in_2023_majority_of_color = IF(D1.`Median medical debt in collections in _2023 - Majority of Color` = 'NA', median_values.median_median_medical_debt_in_2023_majority_of_color, updated_median_medical_debt_in_2023_majority_of_color),
  updated_hospital_market_concentration = IF(D1.`Hospital market concentration _HHI_` = 'NA', median_values.median_hospital_market_concentration, updated_hospital_market_concentration),
  updated_number_of_closures_and_mergers = IF(D1.`Number of Closures and Mergers` = 'NA', median_values.median_number_of_closures_and_mergers, updated_number_of_closures_and_mergers),
  updated_share_of_population_no_health_insurance = IF(D1.`Share of the population with no health insurance coverage` = 'NA', median_values.median_share_of_population_no_health_insurance, updated_share_of_population_no_health_insurance),
  updated_share_nonelderly_adults_disability = IF(D1.`Share of non-elderly adults with a reported disability` = 'NA', median_values.median_share_nonelderly_adults_disability, updated_share_nonelderly_adults_disability),
  updated_average_household_income_2023 = IF(D1.`Average household income in _2023` = 'NA', median_values.median_average_household_income_2023, updated_average_household_income_2023)
FROM `ba775-a07-team.Medical_Debt_Default_Risk_Analysis.median_values` AS median_values
WHERE D1.`State Abbreviation` = median_values.`State Abbreviation`
  AND (
    D1.`Share with medical debt in collections` = 'NA' OR
    D1.`Median medical debt in collections in _2023` = 'NA' OR
    D1.`Share with medical debt in collections - Majority White` = 'NA' OR
    D1.`Median medical debt in collections in _2023 - Majority White` = 'NA' OR
    D1.`Share with medical debt in collections - Majority of Color` = 'NA' OR
    D1.`Median medical debt in collections in _2023 - Majority of Color` = 'NA' OR
    D1.`Hospital market concentration _HHI_` = 'NA' OR
    D1.`Number of Closures and Mergers` = 'NA' OR
    D1.`Share of the population with no health insurance coverage` = 'NA' OR
    D1.`Share of non-elderly adults with a reported disability` = 'NA' OR
    D1.`Average household income in _2023` = 'NA'
  );
```

## 2. Data Analysis

### 2.1 Insurance Status Exploration and Analysis

```python
%%bigquery

SELECT
  `State Abbreviation` AS State,
  AVG(CAST(`Median medical debt in collections in _2023` AS FLOAT64)) AS Avg_Medical_Debt,
  AVG(CAST(`Average household income in _2023` AS FLOAT64)) AS Avg_Income,
  AVG(CAST(`Share of the population with no health insurance coverage` AS FLOAT64)) AS Avg_Uninsured_Rate
FROM `ba775-a07-team.Medical_Debt_Default_Risk_Analysis.FINAL_CLEANED`
GROUP BY `State Abbreviation`
ORDER BY Avg_Medical_Debt DESC
LIMIT 5;
```

The average medical debt in the top 3 states is more than 3 times higher than that of the bottom 3 states, showing stark disparities in financial burden.

The top 3 states have uninsured rates ranging from 12.97% to 21.42%, significantly higher than the bottom 3 states, which have uninsured rates below 7%. This highlights the important role of insurance in reducing medical debt.

The top 3 states are from the West, while the bottom 3 states are in the Northeast and Midwest, reflecting regional differences in healthcare systems, policies, and access.

```python
%%bigquery

-- Step 1: Calculate the approximate median medical debt
WITH MedianDebt AS (
    SELECT
        APPROX_QUANTILES(CAST(`Median medical debt in collections in _2023` AS FLOAT64), 100)[OFFSET(50)] AS MedianDebt2023
    FROM
        `ba775-a07-team.Medical_Debt_Default_Risk_Analysis.FINAL_CLEANED`
)

-- Step 2: Classify states and calculate average uninsured share for each group
SELECT
    CASE
        WHEN CAST(`Median medical debt in collections in _2023` AS FLOAT64) >= (SELECT MedianDebt2023 FROM MedianDebt) THEN 'High Medical Debt'
        ELSE 'Low Medical Debt'
    END AS Debt_Group,
    AVG(CAST(`Share of the population with no health insurance coverage` AS FLOAT64)) AS Avg_Uninsured_Share
FROM
    `ba775-a07-team.Medical_Debt_Default_Risk_Analysis.FINAL_CLEANED`
GROUP BY
    Debt_Group;
```

States were divided into "high medical debt" and "low medical debt" groups. The results showed that the average uninsured rate in the high medical debt group was 13.42%, significantly higher than the 9.98% in the low medical debt group, indicating that lack of health insurance was strongly associated with increased medical debt.
