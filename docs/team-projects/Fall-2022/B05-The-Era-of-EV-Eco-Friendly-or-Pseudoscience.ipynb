<h1>The Era of EV: Eco-Friendly or Pseudoscience?</h1>

#### Project Objective

This project aims to analyze the correlation between the rise of electric vehicles versus CO2 emissions and energy consumption and production which mainly includes electricity from fossil fuels, nuclear, and renewables. Various other factors would also be evaluated to find out if factors like the current level of renewable energy affect the level of EV adoption, or in general, what are some differences made by the rise of EVs or does EVs actually made a difference.

#### Data Sources

1. British Petroleum’s report on energy production/consumption (https://www.bp.com/en/global/corporate/energy-economics/statistical-review-of-world-energy/downloads.html)--- Table 3: Primary Energy consumption per captial, and Table 4: CO2 emissions 
2. International Energy Agency (https://www.iea.org/data-and-statistics/data-tools/global-ev-data-explorer)--- Table 1: EV car sales and Table 2: Electricity mix bar

#### Introduction & motivation

A large portion of the automobile manufacturers start to only produce electric vehicles, and even with California regulators passed rules banning the sale of new gas-powered cars by 2035. Furthermore, You definitely cannot ignore the continued rise of TESLA and all the high-profile tweets from Elon Musk. With all the new high-tech electric vehicles coming out of the market, Nowadays people tend to believe that EVs are more fuel-efficient and sustainable for the environment as compared to traditional internal combustion cars without having substantial evidence in their mind. Thus, we want to ask a question: do EVs in their current state actually help the environment, or are instead detrimental owing to the source of the electricity being used to charge these vehicles?


####  Report Summary

Based on all the findings that we have :

<h3>2) Data preprocessing </h3>

The main purpose of our data preprocessing is to filter out all the useful data for our project objective from the four tables that we have. The EV sales table is our most essential table in this project. As you can see above, there are tables containing data from hundreds of countries and tables that only contain data from 28 countries(EV sales table). Thus, in order to be able to extract findings from our tables, we need to unify all the tables with the EV sales table based on the distinct 28 countries.

#### a. Data cleaning and Unification

<strong>Table 1: EV car sales</strong>
* There are multiple columns that we do not need, and the only columns that we need for our analysis are: country, year, and value of the sales. NOTE: The EV sales table only contains data from 2010 to 2021.

```python
%%bigquery
SELECT region as country,
          year,
          sum(value) as value
  FROM `ba775-team-b5.ba775_dataset.ev_car_sales`
  GROUP BY year, country
LIMIT 3
```

*We found out that there are only 28 distinct countries from the EV sales table. (This query generated 30 country names including "Other Europe" and "World" which are the two we need to get rid of)

```python
%%bigquery
SELECT DISTINCT(country) FROM `ba775-team-b5.ba775_dataset.ev_car_sales_cleaned`
LIMIT 3
```

* Filter out only the Distinct 28 countries from the EV sales table

```python
%%bigquery
SELECT * FROM `ba775-team-b5.ba775_dataset.ev_car_sales_cleaned`
WHERE country in ('Italy', 'Germany', 'United Kingdom', 'Japan', 'China', 'Brazil', 'Spain', 'Denmark', 'Poland', 'Sweden', 'Iceland','South Africa', 'India', 'Portugal', 'Korea','USA' ,'France', 'Switzerland', 'Greece','Norway', 'New Zealand', 'Belgium', 'Netherlands','Chile', 'Finland', 'Mexico', 'Canada', 'Australia')
LIMIT 3
```

<strong>Table 2: Electricity mix bar (TWh)</strong>


* Clean up this table a little bit by cast the datatype and change column names. NOTE: this dataset contains data from 1985 to 2021

```python
%%bigquery
SELECT Entity as country,
      cast(Code as STRING) as code,
      cast(Year as INTEGER) as year,
      cast(Electricity_from_fossil_fuels__TWh_ as FLOAT64) as elec_fossil_fuels,
      cast(Electricity_from_nuclear__TWh_ as FLOAT64) as elec_nuclear,
      cast(Electricity_from_renewables__TWh_ as FLOAT64) as elec_renewables,
FROM `ba775-team-b5.ba775_dataset.elec_mix_bar`
LIMIT 3
```

<strong>Table 3: Primary Energy consumption per captial(Gigajoule per capita)</strong>

* Then, we continued unified consumption per capita table with 28 distinct countries. NOTE: this dataset contains data from 1965 to 2021.

```python
%%bigquery
SELECT * FROM `ba775-team-b5.ba775_dataset.energy_cons_per_capita` 
WHERE country in ('Italy', 'Germany', 'United Kingdom', 'Japan', 'China', 'Brazil', 'Spain', 'Denmark', 'Poland', 'Sweden', 'Iceland','South Africa', 'India', 'Portugal', 'Korea','USA' ,'France', 'Switzerland', 'Greece','Norway', 'New Zealand', 'Belgium', 'Netherlands','Chile', 'Finland', 'Mexico', 'Canada', 'Australia')
LIMIT 3
```

! ! ! NOTICE that this table is a "WIDE" table instead of a "LONG" table(like ev sales table), and all the column name has an underscore sign before all the year columns. In order to be consistent with the time frame and the format of the year and for joining all the tables together later, we need to clean it up by UNPIVOT the table to a "LONG" table and limit it to a time frame from 2010 to 2021.

```python
%%bigquery
SELECT *
FROM `ba775-team-b5.ba775_dataset.28_cons_per_capita_2010to2021`
UNPIVOT
(
  value FOR year IN (growthrate_2021,growthrate_2011_2021,_2010,_2011,_2012,_2013,_2014,_2015,_2016,_2017,_2018,_2019,_2020,_2021)
) AS Unpvt_year
LIMIT 3
```

* ! ! ! We create a value column for all the consumption per capita values of each country of each year, and we create a year column that contains all the year values.

* then we get rid of the underscore sign for the all the values by using UPDATE SET AND SUBSTRING to extract the value we need.

```python
%%bigquery
UPDATE `ba775-team-b5.temp.unpivot cons per capita`
SET year = SUBSTRING(year, 2, 50) 
WHERE year LIKE '_2%'
```

#### b. Joining

Now we finish unified all the tables, it is time to join them together!

```python
%%bigquery
SELECT ABC.country,ABC.year,ABC.ev_sales,ABC.cons_per_capita,ABC.elec_fossil_fuels,ABC.elec_nuclear,ABC.elec_renewables,D.co2_emissions
FROM
(SELECT AB.country,AB.year,AB.ev_sales,AB.cons_per_capita,ROUND(C.elec_fossil_fuels,2)AS elec_fossil_fuels,ROUND(C.elec_nuclear,2)AS elec_nuclear,ROUND(C.elec_renewables,2)AS elec_renewables
FROM
(SELECT A.country,A.year,A.value AS ev_sales,B.cons_per_capita
FROM `ba775-team-b5.ba775_dataset.ev_car_sales_cleaned_unified` AS A
LEFT JOIN
(SELECT country,year,value AS cons_per_capita
FROM `ba775-team-b5.ba775_dataset.unpvt_cons_per_capita_unified`) AS B
ON A.country = B.country AND CAST(A.year AS STRING) = B.year) AB
LEFT JOIN
(SELECT country, year,elec_fossil_fuels,elec_nuclear,elec_renewables
FROM `ba775-team-b5.ba775_dataset.elec_mix_bar_country_unified`) AS C
ON AB.country = C.country AND AB.year = C.year) ABC
LEFT JOIN
(SELECT country, year, value AS co2_emissions
FROM `ba775-team-b5.ba775_dataset.CO2_emissions_from_energy_unified_unpivot`
) D
ON ABC.country = D.country AND CAST(ABC.year AS STRING) = D.year
LIMIT 3

##JOIN ev car sales and cons per capita and elec mix bar and co2 emissions from 2010 to 2021
```

We now have all the ev_sales,cons_per_capita,elec_fossil_fuels,elec_nuclear,elec_renewables,co2_emissions values all in one table with its corresponding year and country. In order to be consistent with the EV sales table, this all in one table constains data from 28 distinct countries from year 2010 to 2021.

The main reason we decide to join all the tables together is for our dashboard creation, and it would be much more convenience to use one table to create a dashboard instead of several tables and also for our audiences to view and understand. However, for answering the questions and generating findings, we use four separate tables since it simplify our process and we are able to use data that are out of the 2010 to 2021 time frame.

<h3>3) Data Evaluation and Analysis</h3>

For the initial evaluation of tables, we want to discover trends and facts from the four tables we have in order to find correlations and generate findings to support or reject our objectives.

Furthermore, from the facts and trends that we discover, we can then find the correlation between facts and trends to generate results followed by our analysis.

<strong> FINDING 1: From 2010 to 2021, With the total EV sales going up drastically in the world, the total fossil fuel energy production is still increasing slightly at a constant rate which means the rise of the EV market does not stop us to burn more fossil fuels to charge our EV vehicles. However, we can see that total renewable energy production doubled which means the world as a whole is focusing on sustainability.
</strong>

* Trend : Total Global EV sales drastically increase each year---which could potentially mean that gasoline vehicles sales go down

```python
%%bigquery
SELECT year, sum(value) as global_ev_sales 
FROM `ba775-team-b5.ba775_dataset.ev_car_sales_cleaned`
GROUP BY year 
ORDER BY year DESC
LIMIT 3
```

* Trend: The three charts below show the total production of fossil fuels, nuclear, and renewable energy in the world(28 distinct countries) in each year from 2010 to 2021. Based on the result, the most produced energy is fossil fuels. For second and third is renewable and nuclear energy.---We can see the total nuclear energy production does not change much, the total fossil fuel production slightly increases, and the total renewable energy production doubles!

```python
%%bigquery
SELECT year, ROUND(SUM(elec_fossil_fuels), 2) AS total_fossil_fuels, ROUND(SUM(elec_fossil_fuels) * 100 / SUM(SUM(elec_renewables+elec_nuclear+elec_fossil_fuels)) OVER (), 2) AS percentage
FROM `ba775-team-b5.ba775_dataset.elec_mix_bar_country_unified`
GROUP BY year
ORDER BY year DESC
LIMIT 5;
```

```python
%%bigquery
SELECT year, ROUND(SUM(elec_nuclear), 2) AS total_nuclear, ROUND(SUM(elec_nuclear) * 100 / SUM(SUM(elec_renewables+elec_nuclear+elec_fossil_fuels)) OVER (), 2) AS percentage
FROM `ba775-team-b5.ba775_dataset.elec_mix_bar_country_unified`
GROUP BY year
ORDER BY year DESC
LIMIT 5;
```

```python
%%bigquery
SELECT year, ROUND(SUM(elec_renewables), 2) AS total_renewables, ROUND(SUM(elec_renewables) * 100 / SUM(SUM(elec_renewables+elec_nuclear+elec_fossil_fuels)) OVER (), 2) AS percentage
FROM `ba775-team-b5.ba775_dataset.elec_mix_bar_country_unified`
GROUP BY year
ORDER BY year DESC
LIMIT 5;
```

<strong> FINDING 2: From 2010 to 2021, The rapid expansion of the EV market in China has also caused China to burn a lot more fossil fuels which further shows the rise of the EV market does not help us produce more energy in an eco-friendly way. However, China has the highest renewable energy production which shows the attitude of pursuing sustainable development.
</strong>

* Fact: The top five countries with the highest total EV sales from 2010 to 2021 are China, USA, Germany, UK, and France.

```python
%%bigquery
SELECT country, SUM(value) AS total_EV_sales, ROUND(SUM(value) * 100 / SUM(SUM(value)) OVER (), 2) AS percentage
FROM `ba775-team-b5.ba775_dataset.ev_car_sales_cleaned_unified`
GROUP BY country
ORDER BY percentage DESC
LIMIT 5
```

* Trend: The three charts below show the total production of fossil fuels, nuclear, and renewable energy of each country from 2010 to 2021.---We can see that from 2010 to 2021, China has the most total fossil fuel energy production followed by the USA. The USA has the most total nuclear energy production and its amount is twice France's total nuclear energy production. China has the most total renewable energy production and its amount is twice of USA's total renewable energy production.

* China has the most total EV sales(47.6%), and it is also the country with the most total renewable energy production. Compare to the USA, the USA has the second total EV sales(14.35%) and second total fossil fuel energy production, and the most nuclear energy production. These facts further support our findings.

```python
%%bigquery
SELECT country, ROUND(SUM(elec_fossil_fuels), 2) AS total_fossil_fuels
FROM `ba775-team-b5.ba775_dataset.elec_mix_bar_country_unified`
WHERE CAST(year AS STRING) IN ('2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021')
GROUP BY country
ORDER BY total_fossil_fuels DESC
LIMIT 3
```

```python
%%bigquery
SELECT country, ROUND(SUM(elec_nuclear), 2) AS total_nuclear
FROM `ba775-team-b5.ba775_dataset.elec_mix_bar_country_unified`
WHERE CAST(year AS STRING) IN ('2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021')
GROUP BY country
ORDER BY total_nuclear DESC
LIMIT 3
```

```python
%%bigquery
SELECT country, ROUND(SUM(elec_renewables), 2) AS total_renewables
FROM `ba775-team-b5.ba775_dataset.elec_mix_bar_country_unified`
WHERE CAST(year AS STRING) IN ('2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021')
GROUP BY country
ORDER BY total_renewables DESC
LIMIT 3
```
